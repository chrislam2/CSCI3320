```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_file = pd.read_csv('random_generated_data.csv')
data_points = data_file[['0', '1']].values
plt.scatter(data_points[:, 0], data_points[:, 1])
plt.xlabel('0')
plt.ylabel('1')
plt.title('Data Points')
plt.show()
```

```python
def K_means(points, k):
    centroids = points[np.random.choice(range(len(points)), size=k, replace=False)]
    
    num_iteration = 0
    while(True):
        num_iteration += 1

        distances = np.sqrt(((points - centroids[:, np.newaxis])**2).sum(axis=2))
        labels = np.argmin(distances, axis=0)
        
        new_centroids = np.array([points[labels == i].mean(axis=0) for i in range(k)])
        
        if np.all(centroids == new_centroids):
            break
        
        centroids = new_centroids
    
    return labels, centroids, num_iteration

labels, centroids, num_iterations = K_means(data_points, 5)

plt.scatter(data_points[:, 0], data_points[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red')
plt.xlabel('0')
plt.ylabel('1')
plt.title('K-means Clustering')
plt.show()

distances = np.sqrt(((data_points - centroids[labels])**2).sum(axis=1))
sum_squared_distances = distances.sum()
print("Number of iterations:", num_iterations)
print("Sum of squared distances:", sum_squared_distances)
```

```python
for k in range(1, 13):
    print(f'--- k = {k} ---')
    labels, centroids, num_iterations = K_means(data_points, k)

    plt.scatter(data_points[:, 0], data_points[:, 1], c=labels)
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red')
    plt.xlabel('0')
    plt.ylabel('1')
    plt.title('K-means Clustering')
    plt.show()

    distances = np.sqrt(((data_points - centroids[labels])**2).sum(axis=1))
    sum_squared_distances = distances.sum()
    print("Number of iterations:", num_iterations)
    print("Sum of squared distances:", sum_squared_distances)
```
